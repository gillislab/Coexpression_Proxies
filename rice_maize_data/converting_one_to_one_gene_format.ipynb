```python
import pandas as pd
```

```python
one_to_one_old_format = pd.read_csv('/data/passala/Data_from_CoCoCoNet/One_to_one_mapping_from_cococonet/arabi_maize_one_one_needs_conversion.csv')
one_to_one_old_format
```

```python
gene_info_maize = pd.read_csv('/data/CoCoCoNet/geneInfo/maize_info.csv')
gene_info_arabi = pd.read_csv('/data/CoCoCoNet/geneInfo/arabidopsis_info.csv')
```

```python
one_to_one_old_format = one_to_one_old_format.merge(right = gene_info_arabi[['OrthoID','LocusTag']], left_on= 'arabidopsis', right_on = 'OrthoID')
one_to_one_old_format.rename(columns = {'arabidopsis':'arabidopsis_OrthoID','maize':'maize_OrthoID','LocusTag':'arabidopsis'}, inplace= True)
one_to_one_old_format.drop(columns = 'OrthoID', inplace= True)
one_to_one_old_format
```

```python
gene_info_maize
```

```python
one_to_one_old_format = one_to_one_old_format.merge(right = gene_info_maize[['OrthoID','EnsemblID']], left_on= 'maize_OrthoID', right_on = 'OrthoID')
one_to_one_old_format = one_to_one_old_format.rename(columns = {'EnsemblID':'maize'})
one_to_one_old_format = one_to_one_old_format.drop(columns = 'OrthoID')
one_to_one_old_format
```

```python
one_to_one_old_format.to_csv('/data/passala/Data_from_CoCoCoNet/One_to_one_mapping_from_cococonet/arabi_maize_one_one_converted_format.csv', index= False)
```
