```python
id_indexer = 0
for gene_pair in maize_rice_map.iterrows():
    
    if gene_pair[1]['Group ID'] == 'Unassigned':
        current_rice_gene = gene_pair[1]['rice']
        current_maize_gene = gene_pair[1]['maize']
        maize_rice_map['Group ID'].loc[(maize_rice_map['rice'] == current_rice_gene) & (maize_rice_map['Group ID'] == 'Unassigned')] = id_indexer
        maize_rice_map['Group ID'].loc[(maize_rice_map['maize'] == current_maize_gene) & (maize_rice_map['Group ID'] == 'Unassigned')] = id_indexer

        all_labeled_groups = maize_rice_map.loc[maize_rice_map['Group ID'] == id_indexer]

        all_labeled_groups_rice_genes = all_labeled_groups['rice'].to_list()
        all_labeled_groups_maize_genes = all_labeled_groups['maize'].to_list()

        maize_rice_map['Group ID'].loc[maize_rice_map['rice'].isin(all_labeled_groups_rice_genes)] = id_indexer
        maize_rice_map['Group ID'].loc[maize_rice_map['maize'].isin(all_labeled_groups_maize_genes)] = id_indexer

        id_indexer += 1
```

```python
all_pairs_to_evaluate_for_functional_conservation = pd.DataFrame(columns = ['Rice','Maize','Group Number'])
for group_number in list(set(maize_rice_map['Group ID'].to_list())):
    current_gene_map = maize_rice_map.loc[maize_rice_map['Group ID'] == group_number]
    list_of_rice_genes_in_group = list(set(current_gene_map['rice'].to_list()))
    list_of_maize_genes_in_group = list(set(current_gene_map['maize'].to_list()))
    all_combo_list_current_genes = itertools.product(list_of_rice_genes_in_group,list_of_maize_genes_in_group)
    all_combo_list_current_genes = list(map(list,all_combo_list_current_genes))
    current_list_of_pairs = pd.DataFrame(all_combo_list_current_genes,columns = ['Rice','Maize'])
    current_list_of_pairs['Group Number'] = group_number
    all_pairs_to_evaluate_for_functional_conservation = all_pairs_to_evaluate_for_functional_conservation.append(current_list_of_pairs)
```

```python
all_pairs_to_evaluate_for_functional_conservation['Rice Score'] = np.nan
all_pairs_to_evaluate_for_functional_conservation['Maize Score'] = np.nan
#all_pairs_to_evaluate_for_functional_conservation.to_csv('/home/passala/git/Coexpressalog_Method_Development/all_pairs_From_maize.csv',index = False)
```

```python
trimmed_rice_cococonet = rice_cococonet[rice_cococonet.columns.intersection(maize_rice_map['rice'].to_list())]
trimmed_rice_cococonet = trimmed_rice_cococonet[trimmed_rice_cococonet.index.isin(maize_rice_map['rice'].to_list())]
double_rice_trimmed_cococonet = trimmed_rice_cococonet[trimmed_rice_cococonet.columns.intersection(maize_rice_map_one_to_one['rice'].to_list())]
double_rice_trimmed_cococonet = double_rice_trimmed_cococonet.replace(1,0)

trimmed_maize_cococonet = maize_cococonet[maize_cococonet.columns.intersection(maize_rice_map['maize'].to_list())]
trimmed_maize_cococonet = trimmed_maize_cococonet[trimmed_maize_cococonet.index.isin(maize_rice_map['maize'].to_list())]
double_maize_trimmed_cococonet = trimmed_maize_cococonet[trimmed_maize_cococonet.columns.intersection(maize_rice_map_one_to_one['maize'].to_list())]
double_maize_trimmed_cococonet = double_maize_trimmed_cococonet.replace(1,0)
```

```python
maize_cococonet_ranked = trimmed_maize_cococonet.rank()
rice_cococonet_ranked = trimmed_rice_cococonet.rank()
```

```python
single_pair_junk_threshold = .7
many_to_many_junk_threshold = .7
difference_between_many_to_drop = .03
### .85 Tolerance = .75,.85,.03
## .7 Tolerance = .7,.8, .04
```

```python
list_of_true_pairs = []
list_of_genes_to_average_and_set_to_be_equal = []
subset_dataframe_length = []
subset_dataframe_width = []

list_of_genes_to_average_and_set_to_be_equal = []
for current_group in list(set(trimmed_all_gene_pairs_for_fc['Group Number'].to_list())):
    dataframe_of_group = trimmed_all_gene_pairs_for_fc.loc[trimmed_all_gene_pairs_for_fc['Group Number'] == current_group]
    dataframe_of_group = dataframe_of_group[['Rice','Maize','Total Score']]
    wide_format = dataframe_of_group.pivot(index = 'Rice',columns= 'Maize',values= 'Total Score')
    if (wide_format>single_pair_junk_threshold).all(axis = None):
        continue
    elif wide_format.shape == (0,0):
        continue
    elif wide_format.shape == (0,1):
        continue 
    elif wide_format.shape == (1,0):
        continue
    elif wide_format.shape == (1,1): # Group is 1 to 1, retain Format: Arabidopsis then Maize
        #one_true_pair = [wide_format.index.item(),wide_format.columns.item()]
        #list_of_true_pairs.append(one_true_pair)
        continue
    elif len(wide_format)>1 and len(wide_format.columns) ==1:
        one_true_pair = [wide_format.idxmin(axis =0).item(),wide_format.columns.item()]
        list_of_true_pairs.append(one_true_pair)
    elif len(wide_format) == 1 and len(wide_format.columns)>1:
        one_true_pair = [wide_format.index.item(), wide_format.idxmin(axis = 1).item()]
        list_of_true_pairs.append(one_true_pair)
    else:

        #Drop Low Quality Columns and Rows
        cols_to_drop = wide_format.columns[wide_format.min()>many_to_many_junk_threshold]
        wide_format = wide_format.drop(columns= cols_to_drop)
        rows_to_drop = wide_format.index[wide_format.min(axis = 1)>many_to_many_junk_threshold]
        wide_format = wide_format.drop(index= rows_to_drop)
        if (wide_format>single_pair_junk_threshold).all(axis = None):
            continue
        elif wide_format.shape ==(0,0):
            continue
        elif wide_format.shape == (1,1): # Group is 1 to 1, retain Format: Arabidopsis then Maize
            one_true_pair = [wide_format.index.item(),wide_format.columns.item()]
            list_of_true_pairs.append(one_true_pair)
        elif len(wide_format)>1 and len(wide_format.columns) ==1:
            one_true_pair = [wide_format.idxmin(axis =0).item(),wide_format.columns.item()]
            list_of_true_pairs.append(one_true_pair)
        elif len(wide_format) == 1 and len(wide_format.columns)>1:
            one_true_pair = [wide_format.index.item(), wide_format.idxmin(axis = 1).item()]
            list_of_true_pairs.append(one_true_pair)
        elif (wide_format>.9).all(axis = None):
            ### Put in retention code here
            both_gene_lists_to_average = [wide_format.index.to_list(), wide_format.columns.to_list()]
            list_of_genes_to_average_and_set_to_be_equal.append(both_gene_lists_to_average)
            
        else:
            for cur_row in wide_format.iterrows():
                cur_row_max = cur_row[1].min()
                cur_row[1][cur_row[1]> cur_row_max + difference_between_many_to_drop] = np.nan
                wide_format.loc[cur_row[0]] = cur_row[1]
            wide_format = wide_format.dropna(axis = 1, how = 'all')
            for cur_col in wide_format.columns:
                cur_col_max = wide_format[cur_col].min()
                wide_format[cur_col].loc[wide_format[cur_col]> cur_col_max+difference_between_many_to_drop] = np.nan
            wide_format = wide_format.dropna(axis = 0, how = 'all')
            col_count = wide_format.count() == 1
            wide_format = wide_format.loc[:,col_count]
            row_count = wide_format.count(axis = 1) ==1 
            wide_format = wide_format.loc[row_count,:]
            wide_format = wide_format.dropna(axis = 1, how = 'all')
            wide_format = wide_format.dropna(axis = 0, how = 'all')
            for label,content in wide_format.items():
                cur_maize_label = label
                cur_arab_label = content.idxmin()
                if type(cur_arab_label) == str:
                    one_true_pair = [cur_arab_label,cur_maize_label]
                    list_of_true_pairs.append(one_true_pair)
            
```

```python
true_pair_dataframe = pd.DataFrame(columns= ['Rice Gene','Maize Gene'], data = list_of_true_pairs)
true_pair_dataframe = true_pair_dataframe.drop_duplicates(subset = 'Rice Gene')
true_pair_dataframe = true_pair_dataframe.drop_duplicates(subset = 'Maize Gene')
true_pair_dataframe
```
